```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.sample(2)
```

```python
credits.sample(2)
```

```python
df=movies.merge(credits,on='title')
df.head(2)
```

```python
d_df=df[['movie_id','title','overview','genres','keywords','cast','crew']]
d_df.head(2)
```

```python
d_df.isnull().sum()
```

```python
d_df.dropna(inplace=True)
```

```python
d_df.duplicated().sum()
```

```python
def convert(obj):
  l=[]
  for i in eval(obj):
    l.append(i['name'])
  return l
```

```python
#d_df['genres']=d_df['genres'].apply(convert)
d_df['keywords']=d_df['keywords'].apply(convert)
```

```python
def convert3(obj):
  l=[]
  counter=0
  for i in eval(obj):
    if counter!=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
d_df['cast']=d_df['cast'].apply(convert3)
```

```python
def fetch_director(obj):
  l=[]
  for i in eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
d_df['crew']=d_df['crew'].apply(fetch_director)
```

```python
d_df['overview']=d_df['overview'].apply(lambda x:x.split())
```

```python
d_df['genres']=d_df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
d_df['keywords']=d_df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
d_df['cast']=d_df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
d_df['crew']=d_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
d_df['tags']=d_df['overview']+d_df['genres']+d_df['keywords']+d_df['cast']+d_df['crew']
```

```python
new_df=d_df[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider')
```
